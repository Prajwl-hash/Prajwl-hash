```python
%%writefile charfreq.l
%{
#include <stdio.h>

int freq[256] = {0};
%}

%%
.|\n    { freq[(unsigned char)yytext[0]]++; }
%%

int main()
{
    yylex();

    printf("\nCharacter Frequencies:\n");

    for(int i = 0; i < 256; i++)
    {
        if(freq[i] > 0)
        {
            if(i == '\n')
                printf("\\n : %d\n", freq[i]);
            else if(i == '\t')
                printf("\\t : %d\n", freq[i]);
            else if(i == ' ')
                printf("' ' : %d\n", freq[i]);
            else
                printf("%c : %d\n", i, freq[i]);
        }
    }

    return 0;
}
```
